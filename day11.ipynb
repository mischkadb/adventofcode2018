```python
import numpy as np
import sys

grid = np.zeros((302,302), dtype=int)
serial_number = 3463

for x in range(1, 301):
    for y in range(1, 301):
        rack_ID = x+10
        power_level = (((((rack_ID * y) + serial_number) * rack_ID) // 100) % 10) - 5
        grid[x,y] = power_level        

def get_largest_total_power(size):
    cur_max = (-sys.maxsize, 0, 0, size)
    for x in range(1, 302-size):
        for y in range(1, 302-size):
            cur_sum = np.sum(grid[x:x+size, y:y+size])
            if cur_sum > cur_max[0]:
                cur_max = (cur_sum, x, y, size)
                
    return cur_max

cur_max = (-sys.maxsize, 0, 0, 0)
for i in range(1, 301):
    cur_val = get_largest_total_power(i)
    if cur_val[0] > cur_max[0]:
        cur_max = cur_val
    print(i, cur_val, cur_max)
    
print(cur_max)
```
